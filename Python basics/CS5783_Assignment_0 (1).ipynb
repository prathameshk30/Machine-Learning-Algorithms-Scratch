```python
# importing the text data
text=open("test.txt","r")
read_text=text.read()
```

```python
#Creating the dictionary of alphabets and their occurences

from collections import OrderedDict
import matplotlib.pyplot as plt

list_text=list(read_text)
counted={}
for i in list_text:
  counted[i]=list_text.count(i)
```

```python
# Generating the graphs for alphabets with separate cases:-
#For each type of case (Upper case/ Lower Case) a different histogram is made to avoid confusion while trying to solve the numpy index question. Check for appropriate graph title.
```

```python
# Histogram for occurence of each Upper Case alphabet :-

keys = ['T', 'G', 'S', 'H','J','P','Q','F','C','A']
counted_capital = {x:counted[x] for x in keys}
list_capital=[key for key, val in counted_capital.items() for _ in range(val)]
plt.figure(figsize=(10,10))
plt.hist(list_capital, bins=20)
plt.title("Distribution of Upper Case alphabet in text")
```

```python
# Histogram for occurence of each lower Case alphabet :-

del counted[',']
del counted[' ']
del counted['\n']
del counted['T']
del counted['G']
del counted['S']
del counted['H']
del counted['J']
del counted['P']
del counted['Q']
del counted['F']
del counted['C']
del counted['A']
del counted['W']

plt.figure(figsize=(10,10))
list_lower=[key for key, val in counted.items() for _ in range(val)]
plt.figure(figsize=(10,10))
plt.hist(list_lower, bins=20)
plt.title("Distribution of Lower Case alphabet in text")
```

```python
# Importing the image

import cv2
img = cv2.imread(r'/content/test.png')
img_1=img
```

```python
# 1) For  each  pixel  in  the  image,  i.e.,  every  element  in  the  array  img, compute the Euclidean distance to the following points [255, 0, 0], [0, 255, 0] and [0, 0, 255]. Once computed, set the value of the pixel to the point with the least distance. Save the image as output1.png 

import numpy as np
list_img=[]
for i in img_1:
  for j in i:
    list_img.append(j)
list_img

list2=([255,0,0],[0,255,0],[0,0,255])
for p,k in enumerate((list_img)):
  list3=[]
  for l,j in enumerate(list2):
    pixel1 = np.array(k)
    pixel2 = np.array(j)
    dist = np.linalg.norm(pixel1 - pixel2)
    list3.append(dist)
    for q,w in zip(list3,list2):
      if q==min(list3):
        list_img[p]=w
        list_img=np.array(list_img)
        list_img.reshape(128,128,3)
print(list_img)
```

```python
import matplotlib.image

matplotlib.image.imsave('output1.png', list_img)
```
